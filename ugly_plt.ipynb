```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

```python
x =[4,2,6,7,9]
y=[1,5,3,8,0]

#resize
plt.figure(figsize=(5,3), dpi=150)

#plt.plot(x,y, label='xy', color='red', linewidth=3, linestyle='--', marker = 'x', markersize=10, markeredgecolor='green')
#shorthand notation
#fmt = '[color][marker][line]'
plt.plot(x,y, 'bo--', label = 'xy')

#line number two
x2 = np.arange(0.4, 4.5, 0.5)
plt.plot(x2[:6], x2[:6] ** 2, 'r-', label='x^2')
plt.plot(x2[5:], x2[5:] ** 2, 'r--', label = 'x^2')

plt.title('1st graph')
plt.xlabel('X!')
plt.ylabel('Y!!')
plt.xticks([0,2,4,6,8,10])
plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```

# Bar chart

```python
labels = ['short', 'tall', 'tallest']
height = [1, 8, 6]

bars = plt.bar(labels, height)

pattern = ['o','-','*']
for bar in bars:
    bar.set_hatch(pattern.pop())
#bars[0].set_hatch('/')
#bars[1].set_hatch('-')
#bars[2].set_hatch('*')

plt.figure(figsize=(6,4))
plt.show()
```
